```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)
plt.plot(np.random.randn(50).cumsum(), 'go--') #if use this always use the last plot
ax1.hist(np.random.randn(100), bins=20, color='#4d0099', alpha=0.3)
ax2.scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30))
ax3.plot(np.random.randn(50).cumsum(), 'go--')
```

```python
fig, axes = plt.subplots(2, 3)
plt.subplots_adjust(wspace=0.5, hspace=0.5) #wspace and hspace is distance between 
axes[0,0].hist(np.random.randn(100), bins=20, color='#4d0099', alpha=0.3)
axes[0,1].plot(np.random.randn(50).cumsum(), 'go--')
axes[0,2].plot(np.random.randn(50).cumsum(), 'go--')
axes[1,0].plot(np.random.randn(50).cumsum(), 'go--')
axes[1,1].plot(np.random.randn(50).cumsum(), 'go--')
axes[1,2].plot(np.random.randn(50).cumsum(), 'go--')
```

```python
fig, axes = plt.subplots(2, 2, sharex=False, sharey=True)
for i in range(2):
    for j in range(2):
        axes[i, j].hist(np.random.randn(500), bins=50, color='k', alpha=0.5)
plt.subplots_adjust(wspace=1, hspace=1)
```

```python
fig = plt.figure(); ax = fig.add_subplot(1, 1, 1)
ax.plot(np.random.randn(1000).cumsum(), 'k', label='one')
ax.plot(np.random.randn(1000).cumsum(), 'k--', label='two')
ax.legend(loc='best')
ax.set_title('My first matplotlib plot')
ax.set_xlabel('Stages')
ax.set_ylabel('Stages')
```

```python
from datetime import datetime
import pandas as pd

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

data = pd.read_csv('data/spx.csv', index_col=0, parse_dates=True)
spx = data['SPX'] # select column 

spx.plot(ax=ax, style='k-')

#prepare the label
crisis_data = [
 (datetime(2007, 10, 11), 'Peak of bull market'),
 (datetime(2008, 3, 12), 'Bear Stearns Fails'),
 (datetime(2008, 9, 15), 'Lehman Bankruptcy')
]

# set the label
for date, label in crisis_data:
 ax.annotate(label, xy=(date, spx.asof(date) + 75),
 xytext=(date, spx.asof(date) + 225),
 arrowprops=dict(facecolor='black', headwidth=4, width=2,
 headlength=4),
 horizontalalignment='left', verticalalignment='top')

#set limit
ax.set_xlim(['1/1/2007', '1/1/2011'])
ax.set_ylim([600, 1800])

#set title
ax.set_title('Important dates in the 2008-2009 financial crisis')
```
